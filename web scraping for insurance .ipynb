```python
result =  requests.get("https://gist.github.com/meperezcuello/82a9f1c1c473d6585e750ad2e3c05a41") 
```

```python
src = result.content
print(src)
```

```python
soup = BeautifulSoup(src, "lxml")
print(soup)
```

```python
ages = soup.find_all("tr", {"class":"js-file-line"})
print(ages)
```

```python
for i in range(len(ages)):
    age.append(ages[i].text)
print(age)
```

```python
insurance1 = [ages[i].getText().strip().split('\n') for i in range(len(ages))]
print(insurance1)
```

```python
df22=pd.DataFrame(insurance1)
df22.to_csv('Medical insurance Data.csv',index=False)
```
